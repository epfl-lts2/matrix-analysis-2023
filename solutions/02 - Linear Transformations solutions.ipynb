### 1. Prove that any set of orthogonal vectors $v_i \in \mathbb{C}^N, \, i=1, \ldots , M \leq N$ such that $v_i^H v_j = C \delta_{i,j}$ is linearly independent (where $C$ is some constant).
---

We have a set of orthogonal vectors $v_i \in \mathbb{C}^N,$  with $i=1, \ldots , M \leq N$.\
Therefore, $v_i^H v_j = C \delta_{i,j} =\left\{\begin{matrix}
 C,& if & i=j\\ 
 0,& if & i\neq j
\end{matrix}\right.$.<br/>

We want to prove that the set is linearly independent, in other words that we have, with $c_{i}$ some scalar, \
$\sum_{j = 1}^{N} c_{j}.v_{j} = 0 \Leftrightarrow c_{j} = 0$.<br/><br/><br/>
Let's start, we can write:\
$v_{i}^H. (\sum_{j = 1}^{N}c_{j}.v_{j}) = v_{i}.0 = 0$<br/><br/>
Therefore,\
$0 = v_{i}^H. (\sum_{j = 1}^{N}c_{j}.v_{j}) =\sum_{j = 1}^{N}c_{j}.v_{j}.v_{i}^H$<br/><br/>
Yet, the set of vectors $v_i$ is orthogonal, $v_i^H v_j = C \delta_{i,j}$. So, all the terms of the sum are equal to zero except for the one where $i=j$:<br/>
$0 =\sum_{j = 1}^{N}c_{j}.v_{j}.v_{i}^H = 0 + c_{i}.v_{i}.v_{i}^H = c_{i}.v_{i}.v_{i}^H$.\
$0 = c_{i}.v_{i}.v_{i}^H$.<br/><br/>
Nonetheless, we know that $v_{i}$ is a non-zero vector, $v_{i}.v_{i}^H =C$. Hence, we can deduce:\
$0 = c_{i}.v_{i}.v_{i}^H$ *if and only if* $c_{i} = 0$<br/><br/>
This reasonning works for all $i=1, \ldots , M \leq N$.<br/><br/>

We can conclude that:\
$\sum_{j = 1}^{N} c_{j}.v_{j} = 0 \Leftrightarrow c_{j} = 0$.<br/>
**The set of orthogonal vectors is linearly independent.**

### 2. Compute $a_r = \sum_{n=0}^{N-1}e^{\frac{2jrn\pi}{N}}$, where $r$ is an integer (discuss the result depending on the value of $r$).
---


If $r\equiv 0 \mod N$, then $b=1$, and $a_r = \sum_{n = 0}^{N-1} 1 = N$.


If $r\not\equiv 0 \mod N$, this is the sum of a geometric series $a_r = \sum_{n = 0}^{N-1} b^{n} = \frac{1 - b^N}{1 - b}$, with $b=e^{\frac{2jr\pi}{N}}$.
Since $b^N = e^{2jr\pi} = 1$, we then have $a_r = 0$.

### 3. Let $v_k \in \mathbb{C}^N$ be such that $v_k[n] = e^{j 2 \pi \frac{kn}{N}}$, for $k,n = 0, \ldots , N-1$. Prove that these vectors are mutually orthogonal, hence linearly independent.
---

$v_{k}$ has the form $\begin{pmatrix}
e^{0}\\ 
e^{j 2 \pi \frac{k}{N}}\\ 
...\\ 
e^{j 2 \pi \frac{kN-1}{N}}
\end{pmatrix}$, for $k,n = 0, \ldots , N-1$.<br/><br/>

As we mentioned in the previous question, to prove that vectors are mutually orthogonal, we need to demonstrate that $v_k^H v_l = C \delta_{k,l}$, with $C$ some constant.<br/><br/>

$v_k^H v_l = \sum_{n = 0}^{N-1} v_{k}[n]^H .v_l[n]$,&nbsp; with $k,l,n = 0, \ldots , N-1$. &nbsp;&nbsp;&nbsp; (simple rule of matrix multiplication)<br/><br/>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $= \sum_{n = 0}^{N-1} e^{-j2\pi k\frac{n}{N}}.e^{j2\pi l\frac{n}{N}} $<br/>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $= \sum_{n = 0}^{N-1} e^{j2\pi \frac{n}{N}(l-k)}$<br/>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $= \sum_{n = 0}^{N-1} e^{j2\pi \frac{n}{N}r}$, &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; if we define $r=l-k$.<br/><br/>

Using the result of question 2, $v_k^H v_l = 0$ for $k\neq l$. 

For $r = 0$, question2 shows that $v_k^H v_l = N$ for $k = l$.<br/><br/>


We can conclude with:<br/>
$v_k \in \mathbb{C}^N$ such as $v_k[n] = e^{j 2 \pi \frac{kn}{N}}$, for $k,n = 0, \ldots , N-1$ give us:<br/>
$v_k^H v_l = N \delta_{k,l}$.


**The vectors are mutually orthogonal, hence linear independent.**

From the above result we have $v_k^Hv_k = N = \|v_k\|^2$, and therefore $\|v_k\| = \sqrt{N}$.

### 4. Let us now consider $c_k\in\mathbb{R}^N$ such that $c_k[n] = \cos\left(\frac{(2n + 1)k\pi}{2N}\right)$. Prove those vectors are also mutually orthogonal. Compute the norm of $c_k$.

Let us try to prove that $c_k^Hc_l = C\delta_{kl}$ for $k,n = 0, \ldots , N-1$ (NB: in this case, the coefficients $c_k[n]$ are real, thus $c_k^H = c_k^T$).

$$
c_k^Hc_l = \sum_{n=0}^{N-1}\cos\left(\frac{(2n + 1)k\pi}{2N}\right)\cos\left(\frac{(2n + 1)l\pi}{2N}\right)
$$

Let us use the trigonometric identity $\cos(a)\cos(b) = \frac{1}{2}(\cos(a+b) + \cos(a-b))$. The above expression becomes
$$
c_k^Hc_l = \frac{1}{2}\sum_{n=0}^{N-1}\cos\left(\frac{(2n + 1)(k+l)\pi}{2N}\right)+ \frac{1}{2}\sum_{n=0}^{N-1}\cos\left(\frac{(2n + 1)(k-l)\pi}{2N}\right)
$$

- The first sum has two cases to consider: 
  - if $k=p=0$ then $k+p=0$. Every term inside the cosine is 0, thus the sum equals $N$.
  - In the other cases, let us rewrite the sum:
$$
\sum_{n=0}^{N-1}\cos\left(\frac{(2n + 1)(k+l)\pi}{2N}\right) = \frac{1}{2}\sum_{n=0}^{N-1}\left(e^{j\frac{(2n + 1)(k+l)\pi}{2N}} + e^{-j\frac{(2n + 1)(k+l)\pi}{2N}}\right) = 
\frac{1}{2}e^{j\frac{(k+l)\pi}{2N}}\sum_{n=0}^{N-1}e^{j\frac{n(k+l)\pi}{N}} + \frac{1}{2}e^{-j\frac{(k+l)\pi}{2N}}\sum_{n=0}^{N-1}e^{-j\frac{(n(k+l)\pi}{N}}
$$

Using the geometric series sum formula, we have $\sum_{n=0}^{N-1}e^{j\frac{n(k+l)\pi}{N}} = \frac{1-e^{j(k+l)\pi}}{1-e^{j\frac{(k+l)\pi}{N}}}$. Similarly, $\sum_{n=0}^{N-1}e^{-j\frac{n(k+l)\pi}{N}} = \frac{1-e^{-j(k+l)\pi}}{1-e^{-j\frac{(k+l)\pi}{N}}}$.

Let us now introduce $b = e^{j\frac{(k+l)\pi}{2N}}$ in the above expressions. We have:
$$
\sum_{n=0}^{N-1}\cos\left(\frac{(2n + 1)(k+l)\pi}{2N}\right) = \frac{1}{2}\left(b\frac{1-b^{2N}}{1-b^2} + b^{-1}\frac{1-b^{-2N}}{1-b^{-2}}\right) = \frac{1}{2}\left(\frac{1-b^{2N}}{b^{-1}-b} + \frac{1-b^{-2N}}{b-b^{-1}}\right)  
= \frac{1}{2}\left(\frac{b^{2N} - 1}{b-b^{-1}} + \frac{1-b^{-2N}}{b - b^{-1}}\right) = \frac{1}{2}\frac{b^{2N}-b^{-2N}}{b - b^{-1}} 
$$
Since $b^{2N} = e^{j(k+l)\pi}$, $b^{2N}-b^{-2N}=2j\sin((l+k)\pi)$, which is 0.

- Now let us consider the second sum
  - if $k=l$, the term inside the cosine is 0, and the sum is $N$.
  - if $k\neq l$:
$$
\sum_{n=0}^{N-1}\cos\left(\frac{(2n + 1)(k-l)\pi}{2N}\right) = \frac{1}{2}\sum_{n=0}^{N-1}\left(e^{j\frac{(2n + 1)(k-l)\pi}{2N}} + e^{-j\frac{(2n + 1)(k-l)\pi}{2N}}\right) = 
\frac{1}{2}e^{j\frac{(k-l)\pi}{2N}}\sum_{n=0}^{N-1}e^{j\frac{n(k-l)\pi}{N}} + \frac{1}{2}e^{-j\frac{(k-l)\pi}{2N}}\sum_{n=0}^{N-1}e^{-j\frac{(n(k-l)\pi}{N}}
$$
Again using the geometric series sum formula, we have 
$$
\sum_{n=0}^{N-1}e^{j\frac{n(k-l)\pi}{N}} = \frac{1-e^{j(k-l)\pi}}{1-e^{j\frac{(k-l)\pi}{N}}},
$$
and
$$
\sum_{n=0}^{N-1}e^{-j\frac{n(k-l)\pi}{N}} = \frac{1-e^{-j(k-l)\pi}}{1-e^{j\frac{-(k-l)\pi}{N}}}.
$$

Let us define $d = e^{j\frac{(k-l)\pi}{2N}}$. We then have:
$$
\sum_{n=0}^{N-1}\cos\left(\frac{(2n + 1)(k-l)\pi}{2N}\right) = \frac{1}{2}d\frac{1-d^{2N}}{1-d^2} + \frac{1}{2}d^{-1}\frac{1-d^{-2N}}{1-d^{-2}}
= \frac{1}{2}\frac{1-d^{2N}}{d^{-1}-d} + \frac{1}{2}\frac{1-d^{-2N}}{d-d^{-1}} = \frac{1}{2}\frac{d^{2N}-d^{-2N}}{d-d^{-1}}
$$

This is very similar to what we got before: $d^{2N}-d^{-2N}=2j\sin((k-l)\pi)=0$.

Combining these results together we get:
- $c_k^Hc_l = N$, if $k=l=0$. Thus, $\|c_0\| = \sqrt{N}$
- $c_k^Hc_l = \frac{N}{2}$, if $k=l\neq 0$. Thus, $\|c_k\| = \sqrt{\frac{N}{2}}$, for $k\neq 0$
- $c_k^Hc_l = 0$, if $k\neq l$

Therefore, those vectors form an orthogonal basis. When doing the normalization, we will have to pay attention to the special case of $c_0$.

### 5. Let us now consider the $N\times N$ Hadamard matrix $H_p$, with $N=2^p$. Prove the matrix is orthogonal and compute the norm of its vectors.
---

First, by construction, it is quite easy to see all $H_p$ matrices are symmetrical, hence $H_p^T=H_p$.
It is quite easy to see that $H_1$ is orthogonal since $H_1H_1^T = H_1H_1 = 2I$.

Let us assume that $H_{p-1}$ is orthogonal and that $H_{p-1}H_{p-1}^T=H_{p-1}H_{p-1}=2^{p-1}I$. Using the definition of $H_p$ we have:
$$
H_pH_p^T = \begin{pmatrix}H_{p-1} & H_{p-1} \\ H_{p-1} & -H_{p-1} \end{pmatrix}\begin{pmatrix}H_{p-1} & H_{p-1} \\ H_{p-1} & -H_{p-1} \end{pmatrix}
$$

Since all matrices composing $H_p$ have the same size, we can block-multiply them easily:
$$
H_pH_p^T= \begin{pmatrix}H_{p-1}H_{p-1} + H_{p-1}H_{p-1}  & H_{p-1}H_{p-1} - H_{p-1}H_{p-1} \\ H_{p-1}H_{p-1} - H_{p-1}H_{p-1} & H_{p-1}H_{p-1} + H_{p-1}H_{p-1} \end{pmatrix}
$$

Non diagonal terms are obviously zero. Diagonal terms are equal to $2H_{p-1}H_{p-1} = 2^pI$.

The norm of each vector is easy to compute, since all elements are either 1 or -1. Computing the scalar product of a vector with itself is just summing 1s so the squared norm of the vector is $N$, and the norm $\sqrt{N}$.

If we reorder the columns (or rows) of the matrix, we do not change the fact that it stays orthogonal. This is true for any orthogonal matrix. We could re-order the DFT matrix (but then would lose the frequency ordering from low to high).

---

```python
import numpy as np
import scipy.misc
import matplotlib.pyplot as plt
```

```python
N = 1024 # you can change this. keep it a power of 2 for the Hadamard matrix.

def get_hadamard_matrix(N):
    nlev = int(np.log2(N)) # number of "levels" to build a hadamard matrix of size N
    if nlev == 1:
        return np.array([[1,1],[1, -1]])
    hm = get_hadamard_matrix(N/2)
    
    return np.block([[hm, hm], [hm, -hm]])

k = np.arange(0, N)
W = np.exp(2j*np.pi*np.outer(k, k)/N)
W_hat = W/np.sqrt(N)

C = np.cos(np.outer(k,2*k+1)*np.pi/(2*N))
C_hat = C*np.sqrt(2)/np.sqrt(N)
C_hat[0, :] = C_hat[0, :]/np.sqrt(2)

H = get_hadamard_matrix(N)
H_hat = H/np.sqrt(N)
```

```python
x_full = scipy.misc.electrocardiogram() 
# the full signal is very long, we will restrict ourselves to 1024 data points to avoid memory issues. You can change the value, keep it a power of 2 to avoid issues with the hadamard matrix
N = 1024 
x_or = x_full[:N]
x = x_or + 0.1*np.random.normal(0, 1, N) # create a noisy version of the signal
plt.plot(x)
plt.plot(x_or, 'r')
```

### 6. Using the normalized versions of the transforms defined and implemented, compute the transform of the noisy signal and plot the result
---

```python
def reverse_Bits(n, no_of_bits):
    result = 0
    for i in range(no_of_bits):
        result <<= 1
        result |= n & 1
        n >>= 1
    return result

def hadamard2walshseq(n):
    b = int(np.log2(n))
    s = np.zeros(n, dtype=int)
    for k in range(n):
        s[k] = reverse_Bits(k^(k>>1), b)
    return s

def hadamard2walsh(H):
    s = hadamard2walshseq(H.shape[0])
    return H[s, :]

Walsh_hat = hadamard2walsh(H_hat)
```

```python
x_dft = W_hat@x
x_dct = C_hat@x
x_walsh = Walsh_hat@x
```

Do the equivalent with the Fourier transform using the first and last $\frac{k}{2}$ coefficients
---

We will keep only $k$ coefficients of the transformed signal, and set the rest to 0.

### Explanation

As students might not (yet) know about the (discrete) Fourier transform, the subject was only asking you to perform this operation without proper explanation. Here is a quick justification of why you would do this.

$\hat{W}x$ is the *Discrete Fourier transform* of the input $x$. Our input signal can be seen as the sampled version of a continuous signal $x(t)$. If we compute the Fourier transform of this continuous input signal we get a transformed signal $\tilde{x}(f)$ that is a function of the frequency, and is a representation of $x$ in the frequency domain.

If you consider a similar operation in the discrete domain (as we did in the exercise), $\hat{W}x$ is a representation of $x$ in the discrete frequency domain. The matrix $\hat{W}$ is used to switch the representation of the input signal from the canonical basis of $\mathbb{R}^N$ to the Fourier basis (as does the Fourier transform in the continuous domain).

The operation suggested to remove the noise is in fact a *low-pass filter*, i.e. a filter that will remove the highest frequencies present in the noisy signal. If we plot the transformed version of both the original and noisy signal, we can see why we try this approach.

Here are some properties of the DFT that we used in this problem:
- the DFT of a signal is periodic with a period $N$
- the DFT of a real signal is symmetrical around 0, i.e. $\hat{W}x (f) = \hat{W}x (-f)$ 

```python
x_dft_or = W_hat@x_or
x_dft = W_hat@x
plt.plot(np.abs(x_dft), alpha=0.7)
plt.plot(np.abs(x_dft_or), 'r', alpha=0.5)
```

The above graph shows the DFT of the original signal (in red) and the noisy one in blue. Remember that the input $x$ is real, therefore has a transform symmetrical around 0. Using the periodicity of the DFT, the "negative frequency" part of the transformed signal is shown between $N/2$ and $N$.

The "high frequency" part lies roughly between 200 and 512 (then is replicated between 512 and 1024-200) and in this part the noisy signal shows a clear difference with respect to the original one: the red curve looks a lot "flatter" than the blue one. The proposed approach aims at "flattening" the high frequency coefficients and remove the noise from the signal.

The DCT and the Walsh transform are also to convert the input signal from a canonical basis to a frequency one (albeit different from the Discrete Fourier transform). The effect of setting high-frequency coefficients in those basis will be similar to the DFT one, i.e. a low-pass filtering operation.

```python
def reconstruct_dct_walsh(input_signal, transform_matrix, k):
    input_signal[k:] = 0
    return transform_matrix.T@input_signal

def reconstruct_dft(input_signal, transform_matrix, k):
    N = len(input_signal)
    input_signal[int(k/2):int(N-k/2)] = 0
    return np.conj(transform_matrix.T)@input_signal # do not forget about complex values

    return None

k = 200 # chose an appropriate value

x_dct_reconstruct = reconstruct_dct_walsh(x_dct, C_hat, k)
x_walsh_reconstruct = reconstruct_dct_walsh(x_walsh, Walsh_hat, k)
x_dft_reconstruct = reconstruct_dft(x_dft, W_hat, k)

```

```python
plt.plot(x_or, 'r')
plt.plot(x_dct_reconstruct, alpha=0.8)
```

```python
plt.plot(x_or, 'r')
plt.plot(np.real(x_dft_reconstruct), alpha=0.8)
```

```python
plt.plot(x_or, 'r')
plt.plot(x_walsh_reconstruct, alpha=0.8)
```

### 8. What would be the advantages/drawbacks of the DCT and Walsh transform over Fourier ?
---

We can see on the reconstructions that the methods seem to perform similarly. They do remove the noise (depending on how many coefficients you keep), but also introduce some distortions: some high frequency information is removed, the amplitude of the peaks is modified, etc. Due to its $\pm 1$ coefficients, the Walsh transform introduces "square" artefacts that might be an issue depending on the application. DFT and DCT show very little difference.

One advantage of DCT over DFT is that it does not involve complex values, hence a simpler implementation. The Walsh transform also avoids the use of complex and its implementation might even be simpler: since the values are only +1 or -1, computing the transform (if we skip the normalization) only involves sign changes and additions. Even in the reconstruction, it will use less multiplications than the DCT, which can make it interesting in constrained computing environments.
